```python
#Criando nosso DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Qtde de linhas e colunas
df.shape
```

```python
df.dtypes
```

```python
#Qual a receita total
df['Valor Venda'].sum()
```

```python
#Qual o custo total
df['custo'] = df['Custo Unitário'].mul(df['Quantidade']) #Criando a coluna de custo de todos os produtos
```

```python
df.head()
```

```python
df['custo'].sum()
```

```python
#Agora que temos a receita e custo e o total, podemos achar o Lucro Total
#Vamos criar uma coluna de lucro que srá Receita - CUsto
df['lucro'] = df['Valor Venda'] - df['custo']
```

```python
df.head()
```

```python
round(df['lucro'].sum(), 2) 
```

### **Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em numérica**

```python
#Criando uma coluna com o total de dias para enviar o produto
```

```python
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head()
```

```python
#Etraindo apenas os dias
df['Tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df['Tempo_envio'].dtype
```

```python
#Média do tempo de envio por marca
df.groupby('Marca')['Tempo_envio'].mean()
```

### **Missing Values**

```python
df.isnull().sum()
```

### **E, se quisermos saber o Lucro por Ano e Por Marca?**

```python
#Agrupar por ano e por marca
pd.options.display.float_format = '{:20,.2f}'.format #formata o número pra ficar no formato correto com 2 casas decimais
df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum()
```

```python
# Resetando o index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
#Gráfico de total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title='Total de produtos vendidos')
plt.xlabel('Total')
plt.ylabel('Produtos');
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['lucro'].sum().plot(title='Lucro x Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
df_2009.groupby('Marca')['lucro'].sum().plot.bar(title='Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby('Classe')['lucro'].sum().plot.bar()
plt.title('Lucro x Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
